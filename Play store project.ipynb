TOP 10 APPS CATEGORIES

```python
Category = pd.DataFrame(ay_copy['Category'].value_counts())
Category.rename(columns={'Category':'Count'},inplace=True)
Category
```

which category has largest number of installation?

```python
df_cat_installs = ay_copy.groupby(['Category'])['Installs'].sum().sort_values(ascending=False).reset_index()
df_cat_installs.Installs = df_cat_installs.Installs/1000000000     # converting a number into billions
df2 = df_cat_installs.head(10)
plt.figure(figsize=(14,10))
sns.set_context("talk")
sns.set_style('darkgrid')

ax = sns.barplot(x='Category', y= 'Installs', data=df2)
ax.set_xlabel('No. of installation in Billions')
ax.set_ylabel('')
plt.xticks(rotation=45)
ax.set_title('Most popural category in Play Store', size = 20)
```

WHAT ARE THE TOP 5 MOST INSTALLED APPS IN EACH POPURAL CATEGORIES?

```python
dfa = ay_copy.groupby(['Category','App'])['Installs'].sum().reset_index()
dfa = dfa.sort_values('Installs', ascending=False)
apps = ['GAME', 'COMMUNICATION','PRODUCTIVITY','SOCIAL']
sns.set_context('poster')
sns.set_style('darkgrid')

plt.figure(figsize=(40,30))

for i,app in enumerate(apps):
    df2 = dfa[dfa.Category == app]
    df3 = df2.head(5)
    plt.subplot(4,2,i+1)
    sns.barplot(data=df3, x='Installs', y='App')
    plt.xlabel('Installation in million')
    plt.ylabel('')
    plt.title(app, size= 20)

plt.tight_layout()
plt.subplots_adjust(hspace= .3)
plt.show()
```

HOW MANY APPS ARE THERE ON GOOGLE PLAY STORE WHICH GET 5 RATINGS?

```python
rating = ay_copy.groupby(['Category','Installs','App'])['Rating'].sum().sort_values(ascending=False).reset_index()
top5rating_apps = rating[rating.Rating==5.0]
print('Number of 5 rated apps',top5rating_apps.shape[0])
top5rating_apps.head(5)
```
